```python
@dataclass 
class Ligne:
    symbole: str
    lien_nyse: str
    securite: str
    lien_wiki: str
    gics: str
    sous_gics: str
    localisation: str
    date_entree: str
    cik: str
    fondation: str
```

```python
def genere_ligne(ligne: str) -> Ligne:
    (
        symbole, 
        securite, 
        gics, 
        sous_gics, 
        localisation, 
        date_entree, 
        cik, 
        fondation
    ) = [td.text.strip() for td in ligne.find_all("td")]
    lien_nyse, lien_wiki, _ = [lien.attrs["href"] for lien in ligne_test.find_all("a")]
    return Ligne(
        symbole=symbole,
        lien_nyse=lien_nyse,
        securite=securite,
        lien_wiki="https://en.wikipedia.org" + lien_wiki,
        gics=gics,
        sous_gics=sous_gics,
        localisation=localisation,
        date_entree=date_entree,
        cik=cik,
        fondation=fondation
    )
```

```python
print(genere_ligne(ligne_test))
```
